```python
def generate_DATA_training(n):
    #% a random training set using the CROSS function
    X = (np.random.rand(n,2)-.5)*2;
    dd=np.transpose(np.array([np.exp(-X[:,0]**2 * 10),np.exp(-X[:,1]**2 * 50),1.25*np.exp(-(X[:,0]**2+X[:,1]**2)*5)]))
    #print(np.exp(-X[:,0]**2 * 10))
    Y = np.transpose([np.max(dd,1)]);
    #print(np.shape(Y))
    Y = Y + np.random.randn(n,1)*0.1;
    #print(np.shape(Y))
    return np.transpose(X),np.transpose(Y)
def generate_DATA_test(n):
    #% a random test set using the CROSS function
    X_list = []
    for i in np.linspace(-1,1,n):
        for j in np.linspace(-1,1,n):
            X_list.append(np.array([i,j]))
    X = np.array(X_list)
    #print(X)
    dd = np.transpose(np.array([np.exp(-X[:,0]**2 * 10),np.exp(-X[:,1]**2 * 50),1.25*np.exp(-(X[:,0]**2+X[:,1]**2)*5)]))
    #print(np.exp(-X[:,0]**2 * 10))
    Y = np.transpose([np.max(dd,1)])    
    #print(np.shape(Y))
    return np.transpose(X),np.transpose(Y)
def makesurf(data,nx):
    # [X,Y,Z]=makesurf(data,nx) converts the 3D data file data into
    # three matices as need by surf(). nx tells how long the row of the
    # output matrices are
    X_ = np.transpose(data[0,:].reshape(nx,nx))
    Y_ = np.transpose(data[1,:].reshape(nx,nx))
    Z_ = np.transpose(data[2,:].reshape(nx,nx))
    
    return X_,Y_,Z_
```

```python
def BLWR_func(action,**kwargs):
    global BLWRS, B
    
    if action == 'Init':
        #ID = kwargs['ID'] # the index showing the local model we are updating (m)
        blwr={}
        blwr['cm'] = kwargs['cm']
        blwr['Lambda'] = kwargs['Lambda']
        blwr['a_alpha'] = kwargs['a_alpha']
        blwr['b_alpha'] = kwargs['b_alpha']
        blwr['a_beta'] = kwargs['a_beta']
        blwr['b_beta'] = kwargs['b_beta']
        blwr['A'] = (blwr['a_alpha']/blwr['b_alpha'])*np.eye(p)
        blwr['beta_hat_fm'] = blwr['a_beta']/blwr['b_beta']
        blwr['a_alpha'] = blwr['a_alpha'] + p/2
        B_elements.append(blwr['beta_hat_fm'])
        B = np.diag(B_elements) # diagonal matrix of precisions
        blwr['mu_w'] = np.zeros([p,1])
        BLWRS.append(blwr)
        
    elif action == 'Change':
        BLWRS
    elif action == 'Update':
        x = kwargs['x']
        #print('x=',x)
        ID = kwargs['ID']
        PHI_ = _phi(x- BLWRS[ID]['cm'],_eta(x, BLWRS[ID]['cm'], BLWRS[ID]['Lambda']))
        AA = BLWRS[ID]['A'] + BLWRS[ID]['beta_hat_fm']*np.dot(np.transpose(PHI_),PHI_)
        #print('AA= ',AA)
        
        BLWRS[ID]['Sigma_w'] = LA.inv(AA)
        #print('Sigma_w=',BLWRS[ID]['Sigma_w'])
        
        BLWRS[ID]['mu_w'] = BLWRS[ID]['beta_hat_fm']* np.dot(BLWRS[ID]['Sigma_w'] , mu_f[ID]*PHI_)
        #print('mu_w= ',BLWRS[ID]['mu_w'] )
        
        #BLWRS[ID]['a_alpha'] = BLWRS[ID]['a_alpha'] + p/2
        #print('a_alpha=',BLWRS[ID]['a_alpha'] )
        
        BLWRS[ID]['b_alpha'] = BLWRS[ID]['b_alpha'] + 0.5*(np.dot(np.transpose(BLWRS[ID]['mu_w']),BLWRS[ID]['mu_w']) 
                                                           + np.trace(BLWRS[ID]['Sigma_w']))
        #print('b_alpha=',BLWRS[ID]['b_alpha'] )
        
        BLWRS[ID]['a_beta'] = BLWRS[ID]['a_beta'] + 0.5
        #print('a_beta=',BLWRS[ID]['a_beta'] )
        
        a1 = mu_f[ID] - np.dot(np.transpose(BLWRS[ID]['mu_w']),PHI_)
        a2 = np.dot(np.transpose(a1),a1)
        a3 = np.dot(np.dot(np.transpose(PHI_),BLWRS[ID]['Sigma_w']),PHI_)
        a4 = 1/BLWRS[ID]['beta_hat_fm'] 
        
        BLWRS[ID]['b_beta'] = BLWRS[ID]['b_beta'] + 0.5*(a2+a3+a4)
        #print(BLWRS[ID]['mu_w'])
        
        BLWRS[ID]['A'] = (BLWRS[ID]['a_alpha']/BLWRS[ID]['b_alpha'])*np.eye(p)
        #print(BLWRS[ID]['A'])
        BLWRS[ID]['beta_hat_fm'] = BLWRS[ID]['a_beta']/BLWRS[ID]['b_beta']
        B[ID,ID] = BLWRS[ID]['beta_hat_fm']
        #print('Lambda=',BLWRS[ID]['Lambda'])
        #print('rate=',LearningRate*_gradient(BLWRS[ID],x))
        BLWRS[ID]['Lambda'] = BLWRS[ID]['Lambda'] + LearningRate*_gradient(BLWRS[ID],x)
    elif action== 'Predict':
        x = kwargs['x']
        
        F=[]
        for m in range(0,M):
            PHI_ = _phi(x-BLWRS[m]['cm'],_eta(x, BLWRS[m]['cm'], BLWRS[m]['Lambda']))
            f_dummy = (np.dot(np.transpose(BLWRS[m]['mu_w']),PHI_))
            F.append(f_dummy)
        #print(F)
        #print(np.shape(np.array(F)))
        return np.squeeze(np.array(F),axis=2)
    elif action == 'Structure':
        BLWRS
    elif action == 'Clear':
        ID = varargin[0]
        del BLWRS[ID]
    return 

def _eta(x, c, Lambda):
    # returns the RBF values for a local model at a given x
    # x belongs to R^d
    if all(item != 0 for item in Lambda):
        LL = np.diag(Lambda);
    #print(LL)
    return np.exp(-0.5 * np.dot(np.dot(np.transpose(x - c), LL**(-1)), (x - c)))[0];
    
def _phi(x,eta):
    epsilon = np.concatenate((np.ones([1,1]),x), axis=0)
    return eta*epsilon
def _gradient(model,x):
    dummy = np.dot(np.transpose(np.power(x - model['cm'], 2)),model['Lambda']**(-2))

    dphi = -0.5*np.dot(_phi(x-model['cm'],_eta(x, model['cm'], model['Lambda'])),dummy)

    grad_ = model['beta_hat_fm'] * np.dot(np.transpose(model['mu_w']), dphi) * (F_n - np.dot(np.transpose(model['mu_w']),_phi(x-model['cm'],_eta(x, model['cm'], model['Lambda']))))
    #print('grad = ',grad_)
    return grad_[0,:]
    
```

```python
##  
BLWRS = []
B_elements=[]
B = 0 # diagonal matrix of precisions
LearningRate = 0.001
p = 2 # number of coefficients in a local linear model
M = 0 # number of local models
w_gen = 0.9
Lambda_init = 0.3**(2)
a0_beta = 50
b0_beta = 10
beta_y = 1000
a0_alpha = 50
b0_alpha = 10
X_train = np.transpose(XX)
Y_train = np.transpose(YY)
d = np.shape(X_train)[0] # input dimension
N = np.shape(X_train)[1] # number of training data samples
inds = np.random.permutation(N)
#print(np.max(inds))
for i in inds: #range(0,N): #Incremental loop (over individual data points)
    ############################  Adding local models #################################
    if M==0:
        #print('M=',M+1)
        #print('i= ', i)
        BLWR_func('Init',ID = 0,cm = np.transpose([X_train[:,i]]), Lambda = Lambda_init*np.ones(d),
                                   a_alpha = a0_alpha, b_alpha = b0_alpha,
                                   a_beta = a0_beta, b_beta = b0_beta)
        M = M + 1
    else:
        dummy =[]
        for m in range(0,M):
            dummy.append(_eta(np.transpose([X_train[:,i]]),BLWRS[m]['cm'],BLWRS[m]['Lambda'])< w_gen)
        if all(item == True for item in dummy):
            #print('M=',M+1)
            #print('i= ', i)
            #print(np.transpose([X_train[:,i]]))
            BLWR_func('Init',ID = M,cm = np.transpose([X_train[:,i]]), Lambda = Lambda_init*np.ones(d),
                                       a_alpha = a0_alpha, b_alpha = b0_alpha,
                                       a_beta = a0_beta, b_beta = b0_beta)
            M = M + 1
    ############################  Updating the local models #################################  
        
    #print('i=',i)
    B_inv = np.diag(np.diag(B)**(-1))
    #print(B_inv,np.trace(B_inv))
    s = 1/beta_y + np.trace(B_inv)

    ## Variational step
    Sigma_f = B_inv - np.dot(np.dot(np.dot(B_inv, np.ones([M,1])), np.ones([1,M])), B_inv)/s
    mu_f = np.dot(Sigma_f,beta_y*Y_train[0,i]*np.ones([M,1])
                      + np.dot(B,BLWR_func('Predict',x = np.transpose([X_train[:,i]]))))
    #print('mu_f = ',mu_f)
    F_n = sum(mu_f)
    #print('F_n=', F_n)
    # Updating. Includes the M-step
    for m in range(0,M):
        #print('m_update=' , m)
        BLWR_func('Update',ID=m,x = np.transpose([X_train[:,i]]))

print('number of local models= ',M)
function_approximation = np.zeros([len(Y_train),1])
    
for i in range(0,len(Y_train)):
    function_approximation[i] = sum(BLWR_func('Predict',x = np.transpose([X_train[:,i]])))
_error = LA.norm(function_approximation-Y_train)/len(Y_train)
print(_error)
```
